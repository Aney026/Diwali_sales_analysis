```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding='latin-1')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.info()
```

```python
# drop unrelated/blank columns

df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
df.shape
```

```python
# change data type
df['Amount']= df['Amount'].astype('int')
```

```python
df.columns
```

```python
# rename column name 
df.rename(columns={'Marital_Status':'Married'})
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
# EDA

ax = sns.countplot(x= "Gender",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y="Amount",data=sales_gen)
```

```python
From above graph we can see that most of the buyers are females.
```

```python
# Age 

ax= sns.countplot(x="Age Group",hue='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount Vs Age group 

sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

from above graph we can see that most of the buyers are of age group b/w 26-35 yrs female 

```python
# Top number of orders from 5 states

sales_state = df.groupby(['State'],as_index=False)["Orders"].sum().sort_values(by='Orders',ascending=False).head(5)
sns.set(rc={'figure.figsize':(8,5)})
sns.barplot(x="State",y="Orders",data=sales_state)
```

```python
sales_state = df.groupby(['State'],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc={'figure.figsize':(8,5)})
sns.barplot(x="State",y="Amount",data=sales_state)
```

```python
# Marital Status 

ax = sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(8,5)})
sns.barplot(x="Marital_Status",y="Amount",data=sales_state,hue='Gender')
```

```python
# Occupation 

sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x='Occupation',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x="Occupation",y="Amount",data=sales_state)
```

from above graph we can see that most of the buyers are working in IT,healthcare and Aviation sector
